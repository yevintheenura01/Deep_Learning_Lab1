```python
exp_data = np.random.exponential(scale=1.0, size=(100000,))
uni_data = np.random.rand(100000)
norm_data = np.random.randn(100000)
```

```python
plt.hist(exp_data, density=True, bins=100, histtype="step", label="Exponential", color="red")
```

```python
plt.hist(uni_data, density=True, bins=100, histtype="step", label="Uniform", color="blue")
```

```python
plt.hist(norm_data, density=True, bins=100, histtype="step", label="Normal", color="green")
```

```python
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

df.head()

```

```python
# Select the specified features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_selected = df[features]

# Calculate Pearson correlation
pearson_corr = df_selected.corr(method='pearson')

# Calculate Spearman correlation
spearman_corr = df_selected.corr(method='spearman')

# Visualize Pearson correlation with heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pearson Correlation Coefficient')
plt.show()

# Visualize Spearman correlation with heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Rank Correlation')
plt.show()
```
